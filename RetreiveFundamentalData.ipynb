```python
def getAssetFundamentalData(assetList):
    #assetList = ["GODREJCP","BAJFINANCE"]
    d = {}
    for asset in assetList:
        url="http://financials.morningstar.com/ajax/exportKR2CSV.html?t="+asset
        data = requests.get(url).content
        rawData = pd.read_csv(io.StringIO(data.decode('utf-8')),sep='delimiter', header=None)
        df = pd.DataFrame()
        df  = rawData[0].str.split(''',(?=(?:[^'"]|'[^']*'|"[^"]*")*$)''',expand=True)
        df.insert(0, 'Asset', asset)
        #print(df.head())
        df.to_csv("fundamental_test.csv", sep=',', encoding='utf-8',mode='a', header=False)
        #d[asset] = df
        #print(asset)
```

```python
def formatHeaderAndTranspose(df,assetList,headerNew):
    df_header=pd.DataFrame(columns=headerNew)
    df_header.loc[0] = headerNew
    df_header.to_csv("fundamental_transpose.csv", sep=',', encoding='utf-8',mode='a', header=False,index=False)
    for asset in assetList:
        df_test = pd.DataFrame()
        df_test = df[df["Asset"] == asset]
        
        
        header = df_test[df_test["Ratio"] == "Efficiency"]
        headers = ["Sequence", "Asset", "Ratio"]
        headerList = [header.iloc[0,i+3].split("-")[0] for i in range(11) ]
        df_test.columns= headers + headerList
        
        df_filtered = df_test[~df_test['TTM'].isin(['TTM'])]
        df_filtered = df_filtered[~df_filtered['TTM'].isin(['Latest Qtr'])]
        df_filtered.fillna(0, inplace=True)

        df_filtered.drop("Asset",axis=1,inplace=True)
        df_filtered.drop("Sequence",axis=1,inplace=True)
        df_filtered = df_filtered.transpose()
        df_filtered.insert(0, 'Asset', asset)
        
        df_filtered.to_csv("fundamental_transpose.csv", sep=',', encoding='utf-8',mode='a', header=False)
```
