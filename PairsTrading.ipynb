```python
import numpy as np
import pandas as pd
from pandas_datareader import data, wb
NSE_SCRIPT_AARTI = data.DataReader(name='AARTIIND.NS', data_source='yahoo',start='2015-01-01')
NSE_SCRIPT_VINATI = data.DataReader(name='VINATIORGA.NS', data_source='yahoo',start='2015-01-01')
NSE_SCRIPT_VINATI.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

NSE_SCRIPT_AARTI['2016']['Close'].plot(figsize=(8, 5))
fig1,ax1 = plt.subplots(1,1,figsize=(12,5))

ax1 = fig1.add_subplot(111)
ax1.plot(NSE_SCRIPT_AARTI['2015':'2017']['Close'],color='b',label='aVal')

ax2 = ax1.twinx()
ax2.plot(NSE_SCRIPT_VINATI['2015':'2017']['Close'],color='g',label='bVal')
```
